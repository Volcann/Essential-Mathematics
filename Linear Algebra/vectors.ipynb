## Vectors
A vector is a one-dimensional array of numbers. It can represent a point in space, a direction, or a set of values in machine learning. Vectors are commonly used to represent features of a data point, where each entry in the vector corresponds to a feature.

````python
import numpy as np
import matplotlib.pyplot as plt

# Vectors
v1 = np.array([2, 3])
v2 = np.array([4, 1])

# Calculate distances
distance_v1_origin = np.linalg.norm(v1)
euclidean_distance = np.linalg.norm(v2 - v1)

# Create the plot
plt.figure(figsize=(10, 8))

# Plot the vector v1 (original vector)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector [2, 3]')

# Plot the vector v2
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector [4, 1]')

# Plot the line connecting v1 and v2
plt.plot([v1[0], v2[0]], [v1[1], v2[1]], 'r--', label='Distance Between [2, 3] and [4, 1]')

# Setting the limits and labels
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')

plt.legend()
plt.title('Distance from Origin and Euclidean Distance')
plt.show()

# Print distances
print(f"Distance of [2, 3] from Origin: {distance_v1_origin:.3f}")
print(f"Euclidean Distance between [2, 3] and [4, 1]: {euclidean_distance:.3f}")
````

1. **NumPy Arrays**:
   - `np.array([2, 3])`: NumPy is a powerful library for numerical computing in Python. In this case, `np.array` is used to create the vector `[2, 3]`. NumPy arrays provide efficient storage and fast mathematical operations on large datasets, making them ideal for handling vectors and matrices in linear algebra.

2. **Scalar Multiplication & Division**:
   - Scalars are single numbers that multiply or divide each element of a vector. In the code, the vector `[2, 3]` is multiplied by `4`, yielding a new vector `[8, 12]`, and divided by `4`, resulting in `[0.5, 0.75]`. These operations demonstrate how scalars change the magnitude of vectors without affecting their direction (except in cases of division which shrinks the vector).

3. **Matplotlib Quiver Plot**:
   - The `quiver` function in `matplotlib.pyplot` is used to plot vectors as arrows. In this case, three arrows represent the original vector, the scalar-multiplied vector, and the scalar-divided vector. The arrows show how the vector changes in direction and magnitude due to scalar operations. The `angles='xy'` and `scale_units='xy'` parameters ensure the vectors are drawn on a consistent scale relative to the plots axes.

4. **Plot Customization (Axes, Gridlines, Legend)**:
   - Several plot customizations are applied to make the graph clearer. `plt.axhline()` and `plt.axvline()` draw horizontal and vertical reference lines along the axes. Gridlines, set using `plt.grid()`, further enhance the visualization by adding dashed lines for better readability. The `plt.legend()` function is used to label the arrows, and `plt.xlim()` and `plt.ylim()` set the range for the x- and y-axes, ensuring the vectors fit neatly into the plot area.

5. **Equal Aspect Ratio**:
   - The line `plt.gca().set_aspect('equal', adjustable='box')` ensures that the scaling of the x- and y-axes is the same, preserving the true angles and lengths of the vectors in the plot. This is crucial for accurately visualizing vector transformations because it ensures the arrows maintain proportional representation. Without this setting, vectors could appear distorted.

````python
import numpy as np
import matplotlib.pyplot as plt

# Original vector
v = np.array([2, 3])

# Scalars
scalar_multiply = 4
scalar_divide = 4

# Vector after scalar multiplication and division
v_multiply = scalar_multiply * v
v_divide = v / scalar_divide

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the original vector
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector [2, 3]')

# Plot the vector after scalar multiplication
plt.quiver(0, 0, v_multiply[0], v_multiply[1], angles='xy', scale_units='xy', scale=1, color='green', label='Scalar Multiply by 4 [8, 12]')

# Plot the vector after scalar division
plt.quiver(0, 0, v_divide[0], v_divide[1], angles='xy', scale_units='xy', scale=1, color='red', label='Scalar Divide by 4 [0.5, 0.75]')

# Setting the limits and labels
plt.xlim(-1, 10)
plt.ylim(-1, 15)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')

plt.legend()
plt.title('Vector Scalar Multiplication and Division')
plt.show()
````

1. **Vector Definition**:
   - `np.array([2, 3])` defines the original 2D vector. NumPy is used to efficiently store and manipulate vectors, where each element represents a component of the vector along the x- and y-axes.

2. **Scalar Addition and Subtraction**:
   - Scalars in this case are added or subtracted element-wise from the vector components. For example, `vector + scalar` adds `2` to both components, resulting in a new vector `[4, 5]`. Similarly, `vector - scalar` subtracts `2` from each component, resulting in `[0, 1]`. These operations shift the vector in the coordinate space.

3. **Plot Customization (Axes, Gridlines, Legend)**:
   - Plot limits are set using `plt.xlim()` and `plt.ylim()` to keep the arrows within a visible range. The dashed lines created by `plt.axhline()` and `plt.axvline()` at `y=0` and `x=0` help to orient the user on the plot. A grid is added for better readability, and `plt.legend()` labels each vector, while `plt.title()` provides context with a title.

4. **Equal Aspect Ratio**:
   - `plt.gca().set_aspect('equal', adjustable='box')` ensures that the x and y dimensions have equal scaling, meaning that the angle and relative length of the vectors are accurate. Without this, the vectors could appear distorted, making the addition and subtraction less visually meaningful.

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the vector and scalar
vector = np.array([2, 3])
scalar = 2

# Create new vectors by adding and subtracting the scalar
vector_plus_scalar = vector + scalar
vector_minus_scalar = vector - scalar

# Plot the original and new vectors
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Original Vector')
plt.quiver(0, 0, vector_plus_scalar[0], vector_plus_scalar[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector + Scalar')
plt.quiver(0, 0, vector_minus_scalar[0], vector_minus_scalar[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector - Scalar')

# Set the plot limits
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Scalar Addition and Subtraction with Vectors')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````

1. **Vector Definition**:
   - `v1 = np.array([2, 3])` and `v2 = np.array([4, 1])` define two 2D vectors using `numpy`. These vectors represent points in a 2D coordinate system, where the first value is the x-component and the second is the y-component.

2. **Vector Addition and Subtraction**:
   - `v_add = v1 + v2` performs vector addition. This operation adds the corresponding components of both vectors, resulting in a new vector `[6, 4]`. Similarly, `v_sub = v1 - v2` performs vector subtraction, yielding `[-2, 2]`. Vector addition combines directions and magnitudes, while subtraction shows the difference in direction and magnitude between two vectors.

3. **Matplotlib Quiver Plot**:
   - `plt.quiver()` is used to plot the vectors as arrows, where each arrow starts at the origin `(0, 0)` and points towards the respective coordinates defined by the vectors. The original vectors `v1` and `v2` are drawn in blue and green, while the result of vector addition is in purple, and subtraction is in red. Each color helps to differentiate between the vectors on the plot.

4. **Plot Customization (Axes, Gridlines, Legend)**:
   - The plot limits are set with `plt.xlim()` and `plt.ylim()` to ensure all vectors are visible within the defined range. The dashed gridlines, created with `plt.grid()`, and axis lines, drawn using `plt.axhline()` and `plt.axvline()`, provide reference points to help visualize the vectors' positions relative to the origin. The `plt.legend()` adds a label to each vector, and `plt.title()` provides a descriptive title for the plot.

5. **Equal Aspect Ratio**:
   - `plt.gca().set_aspect('equal', adjustable='box')` ensures that both x- and y-axes have the same scale, preserving the proportions of the vectors. This helps in accurately visualizing the direction and magnitude of the vectors, as unequal scaling could distort their appearance.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
v1 = np.array([2, 3])
v2 = np.array([4, 1])

# Perform vector addition and subtraction
v_add = v1 + v2
v_sub = v1 - v2

# Create the plot
plt.figure(figsize=(10, 8))

# Plot the original vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector [2, 3]')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector [4, 1]')

# Plot the result of addition
plt.quiver(0, 0, v_add[0], v_add[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Addition [6, 4]')

# Plot the result of subtraction
plt.quiver(0, 0, v_sub[0], v_sub[1], angles='xy', scale_units='xy', scale=1, color='red', label='Subtraction [-2, 2]')

# Setting the limits and labels
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')

plt.legend()
plt.title('Vector Addition and Subtraction')
plt.show()
````

1. **Dot Product**:
   - The **dot product** is a scalar value that is the result of multiplying corresponding components of two vectors and summing them up. The formula for the dot product of vectors `a` and `b` is:
     `dot_product = a_1 . b_1 + a_2 . b_2 + a_3 . b_3`
     In this case, the computation is:
     `1 . 4 + 2 . 5 + 3 . 6 = 4 + 10 + 18 = 32`
     This scalar value, `32`, represents how much the vectors are aligned with each other (larger values indicate stronger alignment, while a dot product of `0` means they are orthogonal).

2. **NumPy's `np.dot()`**:
   - `np.dot(a, b)` is a NumPy function that calculates the dot product of the two vectors. This function efficiently computes the dot product for vectors of any length and can also be used for matrix multiplication when needed.

3. **Output**:
   - The result of the dot product is printed using `print()`, and in this case, the output is:
     ```
     Dot Product: 32
     ```
   This scalar value is useful in various applications such as determining the angle between vectors, projections, and physics calculations like work done by a force.

````python
import numpy as np

# Define the vectors
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Compute the dot product
dot_product = np.dot(a, b)

print("Dot Product:", dot_product)
````

1. **Vector Definition**:
   - `v1 = np.array([1, 0, 0])` and `v2 = np.array([0, 1, 0])` define two basis vectors in a 3D space. These vectors span a 2D subspace (the xy-plane) in 3D. The vector `v1` points along the x-axis, while `v2` points along the y-axis.

2. **Creating a Grid of Points**:
   - The `np.linspace(-1, 1, 10)` function generates 10 evenly spaced points between `-1` and `1` for both the `u` and `v` variables. The `np.meshgrid(u, v)` function creates a grid of points from these `u` and `v` arrays, allowing for the combination of all coordinate pairs.
   - The equations `X`, `Y`, and `Z` compute the coordinates of points in the subspace defined by the vectors. Here, the z-coordinates are `0`.

3. **3D Plotting**:
   - A 3D plot is created using `fig.add_subplot(111, projection='3d')`. This prepares the figure for 3D plotting.
   - The `ax.plot_surface()` function visualizes the 2D subspace by plotting the surface defined by the points `(X, Y, Z)`. The `alpha=0.5` parameter adds transparency to the surface, allowing for better visualization of the origin point and the structure of the plot.

4. **Plotting the Origin**:
   - The origin `(0, 0, 0)` is plotted using `ax.scatter()`, which marks the point with a red color and a size of `100`. This helps to indicate the reference point in the 3D space.

5. **Setting Labels and Legend**:
   - Axis labels are set using `ax.set_xlabel()`, `ax.set_ylabel()`, and `ax.set_zlabel()`, ensuring clarity in the 3D plot. The title is added with `ax.set_title()`, and a legend is included with `ax.legend()` to identify the origin.

6. **Displaying the Plot**:
   - Finally, `plt.show()` displays the 3D plot. The resulting visualization illustrates a 2D plane within the 3D space, demonstrating how two vectors can define a subspace. 

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the vectors spanning the subspace
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])

# Create a grid of points
u = np.linspace(-1, 1, 10)
v = np.linspace(-1, 1, 10)
U, V = np.meshgrid(u, v)
X = U * v1[0] + V * v2[0]
Y = U * v1[1] + V * v2[1]
Z = U * v1[2] + V * v2[2]

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100)

# Plot the origin
ax.scatter(0, 0, 0, color='r', s=100, label='Origin')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('2D Subspace in 3D Space')
ax.legend()

plt.show()
````

This code visualizes two vectors and the subspaces they span in a 2D plane using `numpy` and `matplotlib`. Heres a breakdown of the key concepts used:

1. **Vector Definition**:
   - `v1 = np.array([2, 3])` and `v2 = np.array([-1, 1])` define two 2D vectors. These vectors are used to generate lines that represent the subspaces they span. Each vector represents a direction in the 2D plane, and the subspace spanned by each vector is essentially the line that extends infinitely in both directions along that vector.

2. **Creating Grid of Points**:
   - The `t = np.linspace(-10, 10, 400)` creates 400 evenly spaced points between `-10` and `10`. These points represent the scaling factors applied to the vectors, allowing them to be plotted as lines.
   - For each vector, the x-coordinates (`X1`, `X2`) are represented by `t`, and the y-coordinates are scaled versions of the vector components:
     - For `v1`: `Y1 = v1[0] * t` produces points along the line spanned by `v1`.
     - For `v2`: `Y2 = v2[0] * t` gives the points for the line spanned by `v2`.

3. **Plotting the Subspaces**:
   - `plt.plot(X1, Y1)` and `plt.plot(X2, Y2)` plot the lines representing the subspaces spanned by the vectors `v1` and `v2`. The first line is plotted solid, and the second one is plotted with a dashed style (`linestyle='--'`) to differentiate them visually.

4. **Origin Point**:
   - The origin `(0, 0)` is highlighted using `plt.scatter(0, 0, color='r', s=100)`. This marks the starting point of the vectors in the plane and serves as a visual reference for the subspaces.

5. **Plot Customization (Axes, Gridlines, Legend)**:
   - The x- and y-axes are labeled using `plt.xlabel()` and `plt.ylabel()`. A title is added with `plt.title()` to provide context for the plot.
   - Horizontal and vertical axis lines are drawn using `plt.axhline()` and `plt.axvline()` to mark the x- and y-axes, and `plt.grid()` adds dashed gridlines for better clarity.
   - The `plt.legend()` labels the two subspaces and the origin, making it easy to distinguish between the lines in the plot.

6. **Displaying the Plot**:
   - Finally, `plt.show()` displays the 2D plot, which illustrates how the vectors `v1` and `v2` span their respective subspaces. The solid and dashed lines indicate the directions defined by the two vectors, showcasing how vectors in 2D space can define different directions or subspaces.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors spanning the subspace
v1 = np.array([2, 3])
v2 = np.array([-1, 1])

# Create a grid of points for plotting the subspace
t = np.linspace(-10, 10, 400)
line1 = v1[0] * t
line2 = v2[0] * t

# Compute the points for the lines in the 2D plane
X1 = t
Y1 = line1

X2 = t
Y2 = line2

# Plotting
plt.figure(figsize=(8, 8))

# Plot the lines (subspace) in 2D
plt.plot(X1, Y1, label='Subspace Spanned by v1')
plt.plot(X2, Y2, label='Subspace Spanned by v2', linestyle='--')

# Plot the origin
plt.scatter(0, 0, color='r', s=100, label='Origin')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Subspace in 2D Space')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
````
